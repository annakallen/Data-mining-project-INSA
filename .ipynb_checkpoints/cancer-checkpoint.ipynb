```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.decomposition import PCA
from sklearn import tree
from sklearn import metrics
import pydotplus
from IPython.display import Image
```

```python
data = pd.read_csv("data.csv")
data = data.drop(columns = ['Unnamed: 32', 'id'])
```

```python
data
```

```python
data['diagnosis']=(data['diagnosis']=='M').astype(int)
data.corr()
```

```python
colnames = []
for col in data.columns:
    if col != 'diagnosis':
        colnames.append(col)
features = data[colnames]
data2 = data.copy()
scaler = StandardScaler().fit(features.values)
features = scaler.transform(features.values)
data2[colnames] = features
data2
```

```python
y = data2['diagnosis']
x = data2.drop(columns = ['diagnosis'])

dtree = tree.DecisionTreeClassifier(criterion='entropy',max_depth=7)
dtree = dtree.fit(x, y)

tree_data = tree.export_graphviz(dtree, feature_names=x.columns, class_names=['B', 'M'], filled=True, out_file=None)
tree_graph = pydotplus.graph_from_dot_data(tree_data)
Image(tree_graph.create_png())
```

```python
data3 = data2.sample(50)
y = data3['diagnosis']
x = data3.drop(columns = ['diagnosis'])

dtree = tree.DecisionTreeClassifier(criterion='entropy',max_depth=7)
dtree = dtree.fit(x, y)

tree_data = tree.export_graphviz(dtree, feature_names=x.columns, class_names=['B', 'M'], filled=True, out_file=None)
tree_graph = pydotplus.graph_from_dot_data(tree_data)
Image(tree_graph.create_png())
```

```python
w = data2['diagnosis']
z = data2.drop(columns = ['diagnosis'])
test = dtree.predict(z)
metrics.accuracy_score(test, w, sample_weight=None)
```

```python
data3 = data2.drop(columns = ['diagnosis'])
temp = dtree.predict(data3.to_numpy())
print(temp[15])
print(data2.iloc[15]['diagnosis'])
print(temp[19])
print(data2.iloc[19]['diagnosis'])
```
