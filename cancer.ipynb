```python
old_data = data
data = data.drop(columns = ['fractal_dimension_mean', 'texture_se', 'smoothness_se', 'symmetry_se', 'fractal_dimension_se', 'diagnosis'])
data = data.drop(columns = ['fractal_dimension_worst', 'symmetry_worst', 'smoothness_worst', 'texture_worst', 'symmetry_mean'])
```

```python
from sklearn import cluster
k_means = cluster.KMeans(n_clusters=2, max_iter=20000, random_state=1)
k_means.fit(data)

labels = pd.DataFrame(k_means.labels_,columns=['Cluster ID'])
result = pd.concat((data,labels), axis=1)
# result.plot.scatter(x='radius_worst',y='diagnosis',c='Cluster ID', colormap='jet')
```

```python
centroids = k_means.cluster_centers_
pd.DataFrame(centroids,columns=data.columns)
```

```python
tn = 0
fn = 0
tp = 0
fp = 0
correct = 0

for i in range(1, len(labels)):
    lab = labels.iloc[i, 0]
    truth = old_data.iloc[i, 0]
    if lab == 1 and truth == 1:
        tn += 1
    elif lab == 1 and truth == 0:
        fn += 1
    elif lab == 0 and truth == 0:
        tp += 1
    else:
        fp += 1
    if lab != truth:
        correct += 1
        
print(tn, fn, tp, fp)
print('Recall: ' + str(tp/(tp+fp)))
print('Precision: '+ str(tp/(tp + fn)))
print('F-score: ' + str(2 * (tp/(tp+fp)) * (tp/(tp + fn)) / ((tp/(tp+fp)) + (tp/(tp + fn)))))
```
